```python
import pandas as pd
```

```python
df = pd.read_csv("energy_sample_data.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df["Cost_per_MWh"] = df["Cost_EUR"]/ df["Energy_MWh"]
df
```

```python
avg_cost = df.groupby("Energy_Source")["Cost_per_MWh"].mean()
avg_cost
```

```python
import matplotlib.pyplot as plt

avg_cost.plot(kind="bar")
plt.title("Average Cost per MWh by Energy Source")
plt.xlabel("Energy Source")
plt.ylabel("Cost per MWh (EUR)")
plt.show()
```
